```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

## Part 1: Analyzing Best-Seller Trends Across Product Categories

```python
df['category'].value_counts()
```

```python
df.head()
```

```python


category_crosstab = pd.crosstab(df["category"], df["isBestSeller"])

display(category_crosstab)
```

```python

category_best_seller = (
    df.groupby("category")["isBestSeller"]
    .mean()
    .reset_index()
    .rename(columns={"isBestSeller": "best_seller_proportion"}))


category_best_seller = category_best_seller.sort_values(
    by="best_seller_proportion", ascending=False)

display(category_best_seller)
```

```python
from scipy.stats import chi2_contingency

# Chi-square test for 'category' and 'best_seller'
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(category_crosstab)

chi2_statistic, chi2_p_value
```

- **Analysis:** Given the extremely low p-value, we reject the null hypothesis and suggest that there's a significant association between 'category' and 'best seller'

```python
from scipy.stats.contingency import association

association(category_crosstab, method="cramer")
```

- **Analysis:** Cramer's V value of 0.1223 suggests a weak association between the two variables, though statistically significant.

```python
# Log scale for Y-axis
category_crosstab.T.plot(kind="bar", stacked=True, figsize=(10, 6), legend=False)
plt.yscale("log")  # Apply log scale to reduce extreme differences
plt.xlabel("Best-Seller Status")
plt.ylabel("Log Count")
plt.title("Stacked Bar Chart (Log Scale) of Sale Categories by Best-Seller Status")
plt.tight_layout()
plt.show()
```

## Part 2: Exploring Product Prices and Ratings Across Categories and Brands

```python
# Remove outliers in product prices.

def tukeys_test_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds for the outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify the outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]

    return outliers
```

```python
ds_prices = df['price']

outlier_prices = tukeys_test_outliers(ds_prices)
```

```python
df_cleaned = df[~df["price"].isin(outlier_prices)]
```

```python
sns.boxplot(data=df_cleaned, y='price', palette="coolwarm", legend=False);
plt.show()
```

```python
top_20_categories = df_cleaned["category"].value_counts().nlargest(20).index.tolist()
df_filtered = df_cleaned[df_cleaned["category"].isin(top_20_categories)]

df_top_20_categories = pd.DataFrame(df_filtered)

plt.figure(figsize=(12, 6))
sns.violinplot(
    data=df_top_20_categories, 
    x='category', 
    y='price', 
    palette="coolwarm")
plt.xticks(rotation=45, ha="right") 
plt.tight_layout()
plt.show()
```

```python
df_category_prices = (df_cleaned.groupby("category")["price"].median().reset_index().rename(columns={"price": "median_price"}).sort_values(by="median_price", ascending=False))

display(df_category_prices)
```

**Analysis:** Desktop PCs are the product category with the highest median price.

```python
#Getting top 10 product categories
top_10_categories = df_cleaned["category"].value_counts().nlargest(10).index.tolist()
df_filtered_10 = df_cleaned[df_cleaned["category"].isin(top_10_categories)]
df_top_10_categories = pd.DataFrame(df_filtered_10)
```

```python
df_mean_prices = (df_top_10_categories.groupby("category")["price"].mean().reset_index().rename(columns={"price": "mean_price"}).sort_values(by="mean_price", ascending=False))

plt.figure(figsize=(12, 6))
sns.barplot(data=df_mean_prices, x='category', y='mean_price', hue='category')
plt.xticks(rotation=45, ha="right") 
plt.tight_layout()
plt.show()
```

```python
df_category_prices_mean = (df_cleaned.groupby("category")["price"].mean().reset_index().rename(columns={"price": "mean_price"}).sort_values(by="mean_price", ascending=False))

display(df_category_prices_mean)
```

**Analysis:** Motherboards have the highest average price.

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top_10_categories, x='category', y='stars', palette="coolwarm", legend=False);
plt.xticks(rotation=45, ha="right") 
plt.tight_layout()
plt.show()
```

```python
df_category_stars_median = (df_cleaned.groupby("category")["stars"].median().reset_index().rename(columns={"stars": "median_rating"}).sort_values(by="median_rating", ascending=False))

display(df_category_stars_median)
```

**Analysis:** Computer Memory has the highest median rating at 4.7 stars.
